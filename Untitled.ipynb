```python
m_cleaned = []
for m in m_:
    m = m.replace(" ","")
    m = m[1:] if(m[0] == "0") else m
    m = f"+91{m}" if("+" not in m) else m
    m = m.replace("+910" , "+91")
    m_cleaned.append(m)
```
